```python
#Now going to tune RF and LR to see if we can get better R sqaured and RMSE scores
```

```python
#Model Tuning - RF
#Models
modelRF2 = RandomForestRegressor(n_estimators = 15)

#Model Fitting - Random Forest 2 
modelRF2.fit(X_train, Y_train)
print(cross_val_score(modelRF2, X_train, Y_train))
modelRF2.score(X_train, Y_train)

#Make Predictions - RF 2 
predictionsRF2 = modelRF2.predict(X_test)
predRsquaredRF2 = r2_score(Y_test, predictionsRF2)
rmseRF2 = sqrt(mean_squared_error(Y_test, predictionsRF2))
print('R Squared RF2: %.3f' % predRsquaredRF2)
print('RMSE RF2: %.3f' % rmseRF2)

#Plotting the results - RF 2 
plt.scatter(Y_test, predictionsRF2, color=['blue','green'], alpha = 0.5)
plt.xlabel('Ground Truth')
plt.ylabel('Predictions')
plt.show();
```

```python
modelRF3 = RandomForestRegressor(n_estimators =5)

modelRF3.fit(X_train, Y_train)
print(cross_val_score(modelRF3, X_train, Y_train))
modelRF3.score(X_train, Y_train)

predictionsRF3 = modelRF3.predict(X_test)
predRsquaredRF3 = r2_score(Y_test, predictionsRF3)
rmseRF3 = sqrt(mean_squared_error(Y_test, predictionsRF3))
print('R Squared RF3: %.3f' % predRsquaredRF3)
print('RMSE RF3: %.3f' % rmseRF3)

plt.scatter(Y_test, predictionsRF3, color=['blue','green'], alpha = 0.5)
plt.xlabel('Ground Truth')
plt.ylabel('Predictions')
plt.show();
```

```python
#Model Tuning - LR
#Models
modelLR2 = LinearRegression(normalize = True)

#Model Fitting - LR 2
modelLR2.fit(X_train, Y_train)
print(cross_val_score(modelLR2, X_train, Y_train))
modelLR2.score(X_train, Y_train)

#Make Predictions - LR 2 
predictionsLR2 = modelLR2.predict(X_test)
predRsquaredLR2 = r2_score(Y_test, predictionsLR2)
rmseLR2 = sqrt(mean_squared_error(Y_test, predictionsLR2))
print('R Squared LR2: %.3f' % predRsquaredLR2)
print('RMSE LR2: %.3f' % rmseLR2)

#Plotting the results - LR 2 
plt.scatter(Y_test, predictionsLR2, color=['blue','green'], alpha = 0.5)
plt.xlabel('Ground Truth')
plt.ylabel('Predictions')
plt.show();
```

```python
#Model Tuning - LR
#Models
modelLR3 = LinearRegression(fit_intercept = False)

#Model Fitting - LR 3
modelLR3.fit(X_train, Y_train)
print(cross_val_score(modelLR3, X_train, Y_train))
modelLR3.score(X_train, Y_train)

#Make Predictions - LR 3
predictionsLR3 = modelLR3.predict(X_test)
predRsquaredLR3 = r2_score(Y_test, predictionsLR3)
rmseLR3 = sqrt(mean_squared_error(Y_test, predictionsLR3))
print('R Squared LR3: %.3f' % predRsquaredLR3)
print('RMSE LR3: %.3f' % rmseLR3)

#Plotting the results - LR 3 
plt.scatter(Y_test, predictionsLR3, color=['blue','green'], alpha = 0.5)
plt.xlabel('Ground Truth')
plt.ylabel('Predictions')
plt.show();
```
